# Credits Dashboard

This interactive dashboard helps you visualize and export your credits across time periods.

With this tool you can:
 - Load all your daily aggregated credit data
 - Filter by date range and credit metric
 - Export filtered data to a CSV file placed in your home directory

For more information on Accounts and Credit changes see:
